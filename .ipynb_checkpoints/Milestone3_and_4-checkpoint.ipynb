### Import the necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read the input image and display

```python
img = cv2.imread('img.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig=plt.figure(figsize=(4, 8))
plt.imshow(img)
plt.show()
```

```python
img.shape
```

## Edges

### Milestone 1: Median filter and Edge detection

#### Function to plot two images

```python
def plotImages(im1, im2, t1, t2):
    
    fig=plt.figure(figsize=(10, 14))
    plt.subplot(1, 2, 1)
    plt.title(t1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(im1) 
    plt.subplot(1, 2, 2)
    plt.title(t2)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(im2)  
    plt.show()
```

#### Median filter

Before any further processing, a median filter is applied in order to reduce any salt and pepper noise
that may be present in the input image.

```python
median = cv2.medianBlur(img, 3)
plotImages(img, median, 'Input image', 'Median filtered image')
```

#### Edge detection

For detecting the edges, Canny edge detection technique is used. The benefits of using the Canny edge detector instead of a Laplacian kernel is that the edges are all single pixel edges in the resulting image. This allows for morphological operators to be employed more predictably on the resulting edge image. 

```python
canny_edges = cv2.Canny(median, 230, 50)
plotImages(median, canny_edges, 'Median filtered image', 'Edge detection')
```

### Milestone 2: Morphological operations

#### Dilation

Dilation is performed with a small $2 \times 2$ structuring element. The purpose of this step is to both bolden and smooth the contours of the edges slightly.

```python
kernel = np.ones((2, 2),np.uint8)
dilate = cv2.dilate(canny_edges, kernel, iterations = 1)
plotImages(canny_edges, dilate, 'Canny edge detection', 'Dilation results')
```

### Milestone 3 and 4: Bilateral filter and median filter

```python
bilateral = cv2.bilateralFilter ( img, 5, 50, 50 )
```

#### Downsample the input image by a factor of 4

```python
from skimage.measure import block_reduce

downsampled_image = np.clip(np.round(block_reduce(img, block_size=(4, 4, 1), func=np.mean)), 0, 255).astype(np.uint8)
downsampled_image.shape
```

```python
plotImages(img, downsampled_image, 'Input image', 'Downsampled image')
```

#### Bilateral filtering

```python
bilateral = downsampled_image
for i in range(14):
    bilateral = cv2.bilateralFilter( bilateral, 5, 9, 9 )
```

#### Resize the image

```python
filtered_img = cv2.resize(bilateral, (img.shape[1], img.shape[0]))

plotImages(downsampled_image, bilateral, 'Downsampled image', 'Bilateral filtered image')
```

#### Median filter

```python
median = cv2.medianBlur(img, 3)
plotImages(bilateral, median, 'Bilateral filtered image', 'Median filtered image')
```
